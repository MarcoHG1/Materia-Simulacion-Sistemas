```python
NumerosPseudoAleatorios = []

x1 = 7487
x2 = 8373

multipli = x1 * x2

for i in range(1000):
    cadena = str(multipli)
    while True:
        longitudCadena = len(cadena)
        longitudCadenaInt = int(longitudCadena)
        if longitudCadenaInt == 8:
            extracion4 = cadena[2:6]
            numAleatorios = extracion4
            numAleatorios = '0.' + numAleatorios
            NumerosPseudoAleatorios.append(numAleatorios)
            break
        elif longitudCadenaInt == 4:
            cadena = '0000' + cadena
        elif longitudCadenaInt == 5:
            cadena = '000' + cadena
        elif longitudCadenaInt == 6:
            cadena = '00' + cadena
        elif longitudCadenaInt == 7:
            cadena = '0' + cadena
        elif longitudCadenaInt == 9:
            cadena = cadena[0:8]
        elif longitudCadenaInt == 10:
            cadena = cadena[0:9]

    extracion4Int = int(extracion4)

    multipli = x2 * extracion4Int
    x2 = extracion4Int



print('*-*-*-*')

print('\n'.join(map(str, NumerosPseudoAleatorios)))

print('*-*-*-*-*-*-*-**-')

for i in range(998):
    if NumerosPseudoAleatorios[i] == NumerosPseudoAleatorios[i+1]:
        print(NumerosPseudoAleatorios[i])

#comentado
'''
print('*-*-*-*-*-*-*-**-')
print(multipli)
print(cadena)
print(longitudCadena)
'''

'''
print(type(multipli))
print(type(cadena))
cadena = '0' + cadena
print(cadena)
'''
```
